# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python

plt.figure(figsize=(24,12))
sns.countplot(dataset['Color'],hue=dataset['Buyer Gender'])
plt.title('Bar graph of car colors chosen w.r.t gender',fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Color',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.grid()
```
