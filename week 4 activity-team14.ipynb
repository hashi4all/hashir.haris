```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

# 1) Compare churn count with respect to gender.

```python
plt.figure(figsize=(8,8))
sns.countplot(data['Churn'],hue=data['gender'])
plt.grid()
plt.title('Churn count w.r.t Gender',fontsize=18)


```

# 2)Find out how many female senior citizens there in the dataset 

```python
data[['gender','SeniorCitizen']]
```

```python
femaleSC=len(data[(data['gender']=='Female') & (data['SeniorCitizen']==1) ])
print('Number of female senior citizens =',femaleSC)
```

# 3) Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(10,8))
sns.scatterplot(data['tenure'],data['TotalCharges'],hue=data['Partner'])
plt.title('Comparison Plot of Tenure with Total charge',fontsize=18)
```

```python
#As tenure increases total charges also increases and customer with partner ends up with most expenditure
```

# 4)Find out which contract preffered by the senior citizen. 

```python
month_to_month_preference=len(data[(data['Contract']=='Month-to-month') & (data['SeniorCitizen']==1)])
one_year_preference=len(data[(data['Contract']=='One year') & (data['SeniorCitizen']==1)])
two_year_preference=len(data[(data['Contract']=='Two year') & (data['SeniorCitizen']==1)])

print('Number of senior citizen with month to month prefernce = ',month_to_month_preference)
print('Number of senior citizen with one year prefernce = ',one_year_preference)
print('Number of senior citizen with two year prefernce = ',two_year_preference)
```

```python
#month_to_month contract is prefered by most of the senior citizen
```
